```python
def powerLevel(participant):
    power = 0
    try:
        power += item2value[index2item[participant['stats']['item0']]]
    except:
        power += 0
    try:
        power += item2value[index2item[participant['stats']['item1']]]
    except:
        power += 0
    try:
        power += item2value[index2item[participant['stats']['item2']]]
    except:
        power += 0
    try:
        power += item2value[index2item[participant['stats']['item3']]]
    except:
        power += 0
    try:
        power += item2value[index2item[participant['stats']['item4']]]
    except:
        power += 0
    try:
        power += item2value[index2item[participant['stats']['item5']]]
    except:
        power += 0
    try:
        power += item2value[index2item[participant['stats']['item6']]]
    except:
        power += 0
    return power
```
